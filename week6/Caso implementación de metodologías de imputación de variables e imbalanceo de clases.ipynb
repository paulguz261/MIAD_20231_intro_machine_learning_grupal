### 2. Gestin de datos faltantes (25 puntos)
Realicen una particin de la muestra en un conjunto de entrenamiento y otro de evaluacin. Con la muestra de entrenamiento, realicen un procedimiento de imputacin simple multivariada para todas las variables que tengan datos faltantes en sus filas.

```python
#### Utilicen este espacio para escribir los cdigos del procedimiento del punto 2 ####

XTrain, XTest, yTrain, yTest = train_test_split(data.drop(columns="Class"), data["Class"], test_size=0.2, random_state=7)

imp = IterativeImputer(max_iter=10, random_state=0)
imp.fit(XTrain)

XTrain_imp = pd.DataFrame(imp.transform(XTrain))
XTest_imp = pd.DataFrame(imp.transform(XTest))
```

```python
sns.heatmap(XTrain_imp.isna().T)
```

Como se puede ver, los datos fueron llenados para todas las variables de la data. El mtodo usado fue Imputacin Multivariada, con mximo 10 imputaciones. 

### 3. Estimacin de modelo de clasificacin (10 puntos)
Escojan un modelo de clasificacin de los vistos en la semana 3 del curso e implemntenlo. Evalen su poder predictivo en la muestra de evaluacin debidamente transformada, de acuerdo a la imputacin hecha en el conjunto entrenamiento.

```python
#### Utilicen este espacio para escribir los cdigos del procedimiento del punto 3 ####
from sklearn.metrics import cohen_kappa_score


LDA = LinearDiscriminantAnalysis()

LDA.fit(XTrain_imp, yTrain)
predict = LDA.predict(XTest_imp)
predict_prob = LDA.predict_proba(XTest_imp)

error=0
for i in range(0,len(yTest)):
    if(yTest.iloc[i]!=predict[i]):
        error+=1/len(yTest)
#print('El error de clasificacin para {modelo} en la muestra de test es:{s}'.format(modelo="QDA",s=error))

#print(confusion_matrix(y_true=yTest, y_pred=predict))
accuracy=accuracy_score(y_true=yTest, y_pred=predict)
print("el valor de accuracy es de: {}".format(accuracy))
precisionRegLog= precision_score(y_true=yTest.values, y_pred=predict, average = "macro")
print("el valor de precission es de: {}".format(precisionRegLog))

CKS = cohen_kappa_score(yTest, predict)
print("el valor de Cohen Kappa Score es de: {}".format(CKS))

sns.heatmap(confusion_matrix(y_true=yTest, y_pred=predict), annot=True)
```

En este caso se ve que la clasificacin multivariada es efectiva, ya que la mayora de los valores en la matriz de confusin se ubican en la diagonal, lo cual indica que son predicciones correctas. Tambin se corrobora esto al calcular los valores de *precision* y *accuracy* ya que, al tener valores altos, indican la efectividad del modelo para predecir correctamente las bases. 

Por otro lado, como este es un caso de mltiples clases, se us la mtrica de desempeo Cohen Kappa Score, pues se consider la mtrica ms robusta para multiclasificacin. El valor en este caso es de 0.935, que se considera muy bueno. esto quiere decir que el modelo LDA tiene buen desempeo para la clasificacin.

### 5. Implementacin de QDA (10 puntos)
Realicen una particin de la muestra en un conjunto de entrenamiento y otro de evaluacin. Una vez se ha hecho esto, calibren un modelo QDA con la muestra de entrenamiento y evalen su poder predictivo en la muestra de evaluacin.

```python
#### Utilicen este espacio para escribir los cdigos del procedimiento del punto 5 ####

XTrain, XTest, yTrain, yTest = train_test_split(raw_df.drop(columns="Class"), raw_df["Class"], test_size=0.2, random_state=3)

QDA = QuadraticDiscriminantAnalysis()
QDA.fit(XTrain, yTrain)

predict = QDA.predict(XTest)
predict_prob = QDA.predict_proba(XTest)


error=0
for i in range(0,len(yTest)):
    if(yTest.iloc[i]!=predict[i]):
        error+=1/len(yTest)
print('El error de clasificacin para {modelo} en la muestra de test es:{s}'.format(modelo="QDA",s=error))

print(confusion_matrix(y_true=yTest, y_pred=predict))
accuracyQDA=accuracy_score(y_true=yTest, y_pred=predict)
print("el valor de accuracy es de: {}".format(accuracyQDA))
precisionQDA= precision_score(y_true=yTest, y_pred=predict)
print("el valor de precision es de: {}".format(precisionQDA))

# curva roc para Train
fig, ax = plt.subplots(ncols=2, figsize=(16, 8))
fpr, tpr, thresholds = roc_curve(yTrain, QDA.predict(XTrain))
AUC_trainQDA = auc(fpr, tpr)
ax[0].set_title('Curva ROC para el entrenamiento')
ax[0].plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % AUC_trainQDA)
ax[0].legend(loc = 'lower right')
ax[0].plot([0, 1], [0, 1],'r--')
ax[0].axis(xmin=-0.01,xmax=1,ymin=0,ymax=1.05)
ax[0].set_ylabel('True Positive Rate')
ax[0].set_xlabel('False Positive Rate')

# curva roc para Test
fpr, tpr, thresholds = roc_curve(yTest, predict)
AUC_testQDA = auc(fpr, tpr)
ax[1].set_title('Curva ROC para el conjunto de prueba')
ax[1].plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % AUC_testQDA)
ax[1].legend(loc = 'lower right')
ax[1].plot([0, 1], [0, 1],'r--')
ax[1].axis(xmin=-0.01,xmax=1,ymin=0,ymax=1.05)
ax[1].set_ylabel('True Positive Rate')
ax[1].set_xlabel('False Positive Rate')
```

El modelo tiene una mtrica de *AUC = 0.91* lo cual indica un desempeo favorable del modelo. Sin embargo, al ver los valores de la variable *precision* la cual indica la proporcin de predicciones correctas realizadas sobre el total de predicciones, se evidencia que menos del 60% de las predicciones que realiza el modelo son correctas. Dependiendo del contexto del modelo este puede ser un resultado desfavorable o aceptable. Esta precisin sera desfavorable en un contexto en el cual realizar una prediccin errnea tenga un alto costo.

### 6. Gestin de imbalanceo de clases (25 puntos)
Realicen un procedimiento para gestionar las clases imbalanceadas de la base de datos con el conjunto de entrenamiento. 

```python
#### Utilicen este espacio para escribir los cdigos del procedimiento del punto 6 ####

sm = SMOTE(random_state=7,sampling_strategy=0.35)


XTrain_res, yTrain_res = sm.fit_resample(XTrain, yTrain)
print(len(yTrain_res)/(len(yTrain)-len(yTrain_res)))

gr = pd.concat([XTrain_res,yTrain_res],axis=1)
gr = gr.groupby("Class")[["V1"]].count()
gr["pct"] = gr["V1"]/gr["V1"].sum()
gr
```

Se puede ver que el rebalanceo de las clases para este caso causa un aumento del 100% en los datos de la clase minoritaria. 

### 7. Implementacin de random forest (10 puntos)
Estimen un modelo random forest para predecir la variable de respuesta y midan su desempeo en la muestra de evaluacin.

```python
#### Utilicen este espacio para escribir los cdigos del procedimiento del punto 7 ####

RandomForest = RandomForestClassifier(random_state=3)
RandomForest.fit(XTrain_res, yTrain_res)

prediccionRF=RandomForest.predict_proba(XTest)[:,0]
fpr, tpr, thresholds = roc_curve(yTest, prediccionRF, pos_label=0)
AUCRF = auc(fpr, tpr)
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % AUCRF)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([-0.01, 1])
plt.ylim([0, 1.05])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
predict = RandomForest.predict(XTest)

print(confusion_matrix(y_true=yTest, y_pred=predict))
accuracyRF = accuracy_score(y_true=yTest, y_pred=predict)
print("el valor de accuracy es de: {}".format(accuracyRF))
precisionRF = precision_score(y_true=yTest.values, y_pred=predict)
print("el valor de precission es de: {}".format(precisionRF))
```

En este caso se ve una mejora en las metricas evaluadas tanto el *AUC* y *precission*. En **ste** caso, este modelo es un mejor clasificador que el modleo QDA calculado anteriormente.
