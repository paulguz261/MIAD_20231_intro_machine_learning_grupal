## Pregunta de negocio

**¿Por qué es importante predecir el precio de un nuevo inmueble de airbnb?**

Para los nuevos hosts es de vital importancia la definición de un precio adecuado para su inmueble, a partir de este se determina el éxito o el fracaso de la oferta. Si es la primera vez que se va a ofertar un inmueble el propietario se enfrenta al desafío de establecer un precio adecuado para su propiedad.

tanto así que airbnb dedica parte de sus blogs a orientar a los host para determinar el precio a convenir. 
https://www.airbnb.ca/resources/hosting-homes/a/is-my-space-a-good-fit-for-airbnb-4
https://www.airbnb.ca/help/topic/1475
https://www.airbnb.ca/help/article/52
https://www.airbnb.ca/resources/hosting-homes/g/setting-up-a-successful-listing-3


Entre las sugerencias que da airbnb está el conocer su área y las características de otros inmuebles aledaños, por lo tanto generar un modelo que de forma automática tenga en cuenta estas variables tiene sentido.

Si se ayuda a automatizar dicho proceso se puede mejorar la experiencia para el nuevo host, ya que con un precio adecuado el inmueble estará continuamente en renta lo que generará que el host quiera permanecer en la plataforma, asimismo esta consulta debe poderla realizar en cualquier momento para así actualizar sus precios y así mejorar sus ganancias. También si el modelo es altamente interpretable este ayudará a los host para mejorar sus inmuebles ya que podría incluir servicios que incrementen su valor.


Por otro lado, una herramienta de predicción del precio de renta también puede ser útil para un cliente potencial que tenga una propiedad. Si esta persona quiere entrar al negocio de AirBNB, le sería útil conocer de antemano un precio tentativo con el que puede alquilar su inmueble, o las oportunidades de mejora para optimizar el precio de alquiler. De igual manera, poder monitorear el precio sugerido por el modelo le permite mantenerse competitivo en el mercado.



## Objetivo del proyecto

El propósito es poder pronosticar el **valor diario** que cuesta una vivienda rentada por Airbnb en la ciudad de San Diego, California.

```python
# Crear funciones
from matplotlib.pyplot import figure


def contar_vacios_df(df):
    vacios = df[df.isna()]
    return vacios.shape[0]

#contar_vacios_df(df)

def vacios_por_campo_df(df):
    campos = df.columns
    conteo = {}
    for campo in campos:
        conteo[campo]= sum(df[campo].isna())
    return pd.DataFrame.from_dict(conteo, orient='index').rename(columns={0:f"Registros nulos"})

#vacios_por_campo_df(df)

#fig, axes = plt.subplots(2,2, figsize=(12,12))

def histograma_axes(fig, axes, df, variable, intervalos=100):    
    axes[0,0].hist(x=df[variable], bins=intervalos, color='#F2AB6D', rwidth=0.85)
    axes[0,0].set_title('Histograma')
    axes[0,0].set_xlabel(f'{variable}')
    axes[0,0].set_ylabel('Frecuencia')

#histograma(fig, axes, df, 'Life expectancy ', intervalos=10)

def histograma(df, variable, intervalos=100):
    figure(figsize=(8, 6), dpi=80)
    plt.hist(x=df[variable], bins=intervalos, color='#F2AB6D', rwidth=0.85)
    plt.title('Histograma')
    plt.xlabel(f'{variable}')
    plt.ylabel('Frecuencia')
    plt.show()
#histograma(df, 'Life expectancy ', intervalos=100)

def diagrama_caja(df, continua, agrupacion):
    figure(figsize=(8, 6), dpi=80)
    df[agrupacion] = df[agrupacion].astype(object)
    sns.boxplot(data=df, x=continua, y=agrupacion) #, order=["yes", "no"]
    plt.show()

def correlacion(ejex, ejey, df):
    figure(figsize=(8, 6), dpi=80)
    corr = df[ejex].corr(df[ejey])
    plt.scatter(df[ejex], df[ejey])
    plt.title(f"Dispesion {ejex} vs {ejey}")
    plt.xlabel(f"{ejex}")
    plt.ylabel(f"{ejey}")
    plt.text(max(df[ejex]),max(df[ejey]),f"Correlación: {round(corr,2)}")
    plt.show()

#correlacion(ejex="Adult Mortality", ejey='Life expectancy ', df = df)
# df['Status'].dtype == 'O'
def graficas_dataframe(df, variable_objetivo = 'Life expectancy '):

    for campo in df.columns:
        if campo == variable_objetivo:
            histograma(df = df, variable = campo, intervalos=100)
        elif df[campo].dtype == 'O':
            diagrama_caja(df = df, continua = variable_objetivo, agrupacion = campo)
        else:
            correlacion(ejex = campo, ejey = variable_objetivo, df=df)
#graficas_dataframe(df = df, variable_objetivo = 'Life expectancy ')


from statistics import *

def separar_extremos(df,variable,corte_desviaciones=3):
    mediana = statistics.median(df[variable])
    desviacion = statistics.stdev(df[variable])
    df_sin_extremos = df[df[variable].between(mediana - desviacion*(3), mediana + desviacion*3)]
    df_extremos = df[~df[variable].between(mediana - desviacion*(3), mediana + desviacion*3)]
    return df_sin_extremos, df_extremos

def correlacion_grupos(ejex, ejey, df, hue1=False):
    figure(figsize=(8, 6), dpi=80)
    corr = df[ejex].corr(df[ejey])
    sns.scatterplot(x=df[ejex], y=df[ejey], hue=df[hue1], legend = 'auto');
    plt.title(f"Dispesion {ejex} vs {ejey}")
    plt.xlabel(f"{ejex}")
    plt.ylabel(f"{ejey}")
    plt.text(max(df[ejex]),max(df[ejey]),f"Correlación: {round(corr,2)}")
    plt.show()
    

def cortes_percentiles(df, variable):
    cortes = [0.1,0.25,0.5,0.75,0.9,0.99]
    percentiles = {}
    for i in cortes:
        percentiles[i]=df[variable].quantile(i)
    
    return pd.DataFrame.from_dict(percentiles,orient='index').rename(columns={0:"Valor del percentil"})
```

## bathrooms_text


| Field | Type | Description |
| --- | --- | --- |
| bathrooms_text | string | The number of bathrooms in the listing. On the Airbnb web-site, the bathrooms field has evolved from a number to a textual description. For older scrapes, bathrooms is used. |



```python
def extraer_numero(ele:str):
    patron_numero = re.compile(r"([0-9\.]+)")    
    if str(ele) == 'nan':
        extraccion = np.nan
    elif re.search(r'Half-bath', ele):
        extraccion = 0.5
    elif re.search(r'Shared half-bath', ele):
        extraccion = 1
    else:
        extraccion = float(re.search(patron_numero, ele).group(0))
    return extraccion

df['bathrooms_num'] = df['bathrooms_text'].apply(extraer_numero)
df_sin_ex_price, df_extremos_price = separar_extremos(df,'price')

histograma(df = df , variable = 'bathrooms_num', intervalos=100)
display(cortes_percentiles(df_sin_ex_price, 'bathrooms_num'))
correlacion('bathrooms_num', 'price', df=df_sin_ex_price)

correlacion_grupos(ejex='bathrooms_num', ejey='price', df=df_sin_ex_price, hue1='room_type')

conclusion = "Al parecer no hay diferencia entre la correlación que existe entre el número de camas, habitaciones y baños frente al precio. Es decir, el número de camas y habitaciones aportan la misma información"
print(conclusion)

variables_para_analizar.append('bathrooms_num')
```

## Modelo de ML - pipeline



```python

# Librerias y funciones


from sklearn.dummy import DummyRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_absolute_error, median_absolute_error, roc_auc_score, r2_score, explained_variance_score




def entrenar_pipeline(X_train, y_train, params, k,seed = 125):                 
  pca = PCA(random_state = seed)
  svc = SVC( random_state = seed)
  pipe = Pipeline(steps=[('pca', pca), ('svc', svc)])
  gsearch = GridSearchCV(pipe,   params,cv=k,verbose=0)  
  gsearch = gsearch.fit(X_train, y_train)
  return gsearch




def regression_roc_auc_score(y_true, y_pred, num_rounds = 10000):
  """
  # Funcion tomada en: https://towardsdatascience.com/how-to-calculate-roc-auc-score-for-regression-models-c0be4fdf76bb
  Computes Regression-ROC-AUC-score.
  
  Parameters:
  ----------
  y_true: array-like of shape (n_samples,). Binary or continuous target variable.
  y_pred: array-like of shape (n_samples,). Target scores.
  num_rounds: int or string. If integer, number of random pairs of observations. 
              If string, 'exact', all possible pairs of observations will be evaluated.
  
  Returns:
  -------
  rroc: float. Regression-ROC-AUC-score.
  """
    
    
  y_true = np.array(y_true)
  y_pred = np.array(y_pred)

  num_pairs = 0
  num_same_sign = 0
  
  for i, j in _yield_pairs(y_true, num_rounds):
    diff_true = y_true[i] - y_true[j]
    diff_score = y_pred[i] - y_pred[j]
    if diff_true * diff_score > 0:
      num_same_sign += 1
    elif diff_score == 0:
      num_same_sign += .5
    num_pairs += 1
      
  return num_same_sign / num_pairs


def _yield_pairs(y_true, num_rounds):
  """
  # Funcion tomada en: https://towardsdatascience.com/how-to-calculate-roc-auc-score-for-regression-models-c0be4fdf76bb
  Returns pairs of valid indices. Indices must belong to observations having different values.
  
  Parameters:
  ----------
  y_true: array-like of shape (n_samples,). Binary or continuous target variable.
  num_rounds: int or string. If integer, number of random pairs of observations to return. 
              If string, 'exact', all possible pairs of observations will be returned.
  
  Yields:
  -------
  i, j: tuple of int of shape (2,). Indices referred to a pair of samples.
  
  """
  import numpy as np
  
  if num_rounds == 'exact':
    for i in range(len(y_true)):
      for j in np.where((y_true != y_true[i]) & (np.arange(len(y_true)) > i))[0]:
        yield i, j     
  else:
    for r in range(num_rounds):
      i = np.random.choice(range(len(y_true)))
      j = np.random.choice(np.where(y_true != y_true[i])[0])
      yield i, j
```

```python
# VIF dataframe
from statsmodels.stats.outliers_influence import variance_inflation_factor

## Para no sacrificar poder de explicacion se hace un analisis de VIF, hasta que este sea menor a 5, que es el umbral en que coinciden varios autores
def solucionar_VIF(df):
     # No se puede ingresar un data frame con missing
    eliminar = []
    factor_inflacion = 10
    while factor_inflacion >=5:        
        vif_data = df.dropna().drop(columns=eliminar)
        # calculating VIF for each feature
        vif = pd.DataFrame()
        vif["VIF Factor"] = [variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])]
        vif["features"] = vif_data.columns
        display(vif)
        fila_para_eliminar = vif[vif["VIF Factor"]==vif["VIF Factor"].max()]
        display(fila_para_eliminar)
        mask = np.triu(np.ones_like(vif_data.corr(), dtype=bool))
        sns.heatmap(vif_data.corr(), annot=True, mask=mask, vmin=-1, vmax=1)
        plt.title('Correlation Coefficient Of Predictors')
        plt.show()
        eliminar.append(fila_para_eliminar.iloc[0,1])
        factor_inflacion = fila_para_eliminar.iloc[0,0]
        return vif_data

vif_data = solucionar_VIF(prueba)
```
